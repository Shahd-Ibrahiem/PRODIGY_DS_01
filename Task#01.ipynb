#Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Read The Population Data Set

```python
df = pd.read_csv("Population.csv")
print(df.shape)
```

```python
df.info()
```

```python
df
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["Yearly Change"]=df["Yearly Change"].str.replace('%',"").astype(float)
df["Yearly Change"]
```

```python
df_subset = df[["Country (or dependency)" , "Population (2020)" , "Yearly Change" , "Med. Age" , "Land Area (Km²)"]]
df_subset
```

```python
print(df_subset.sort_values("Population (2020)" , ascending=False))
```

```python
print(df_subset.sort_values("Yearly Change" , ascending=False))
```

```python
min_Population_2020=df_subset["Population (2020)"].min()
min_Population_2020
```

```python
max_Population_2020=df_subset["Population (2020)"].max()
max_Population_2020
```

```python
max_Yearly_Change = df_subset["Yearly Change"].max()
max_Yearly_Change
```

```python
min_Yearly_Change = df_subset["Yearly Change"].min()
min_Yearly_Change
```

```python
df_subset[df_subset["Yearly Change"]==max_Yearly_Change]
```

```python
df_subset[df_subset["Yearly Change"]==min_Yearly_Change]
```

```python
df_subset[df_subset["Population (2020)"]==max_Population_2020]
```

```python
df_subset[df_subset["Population (2020)"]==min_Population_2020]
```

```python
#Visualizing Data
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.bar(df_subset["Yearly Change"],df_subset["Med. Age"],width=.05);
plt.xlabel("Yearly Change")
plt.ylabel("Med. Age")
plt.title("Yearly change in Median Age");
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.hist(df_subset["Med. Age"]);
plt.xlabel("Med. Age")
plt.ylabel("Frequency")
plt.title(" Distribution the Median Age")
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.bar(df_subset["Med. Age"],df_subset["Land Area (Km²)"])
plt.xlabel("Med. Age")
plt.ylabel("Land Area (Km²)")
plt.title("Median Age in Land Area (Km²)");
```
